```python
!pip install mlxtend -q

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
dataset = pd.read_excel("/kaggle/input/datasets/ayahesham2002/market-basket-analysis/Egyptian_Grocery_Transactions.xlsx",header=None)
```

```python
dataset.columns=['items']
dataset
```

```python
#List of lists for encoding 
items=dataset['items'].values
transactions=[]
for i in range (0,len(items)):
    transactions.append(items[i].split(","))
transactions
```

```python
#Hot encoding them 
encoded_data = TransactionEncoder()
ZEROONES = encoded_data.fit_transform(transactions)
ZEROONES
```

```python
encoded_data.columns_
```

```python
Ready_Data=pd.DataFrame(data=ZEROONES,columns=encoded_data.columns_,dtype=int)
```

```python
# Apply Apriori algorithm to find frequent itemsets with at least 10% support,
# then sort results by support in descending order to see the most frequent itemsets first
support = apriori(Ready_Data, min_support=0.1, use_colnames=True)
support = support.sort_values('support', ascending=False)
support
```

```python
# Generate association rules from the frequent itemsets using confidence as the metric,
# keeping only rules with at least 30% confidence (meaning the consequent appears in
# at least 30% of transactions that contain the antecedent),
# then sort by confidence descending to see the strongest rules first
confidence=association_rules(support,metric='confidence',min_threshold=0.3)
confidence.sort_values(by='confidence',ascending=False)
```
